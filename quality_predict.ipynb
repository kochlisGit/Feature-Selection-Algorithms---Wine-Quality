```python
from ipywidgets.widgets import interactive, Dropdown, FloatSlider, IntSlider
from IPython.display import display
import joblib
import numpy as np

regressor = joblib.load('Wine-Preference-Analysis/checkpoints/xgboost_regressor.ckp')


def predict(
        wine_type: str,
        fixed_acidity,
        volatile_acidity,
        citric_acid,
        residual_sugar,
        chlorides,
        free_sulfur_dioxide,
        other_sulfur_dioxide,
        density,
        ph,
        sulphates,
        alcohol
):
    wine_type_value = 0.0 if wine_type == 'White' else 1.0
    sample = np.float64([[
        wine_type_value, fixed_acidity, volatile_acidity, citric_acid, residual_sugar, chlorides,
        free_sulfur_dioxide, other_sulfur_dioxide, density, ph, sulphates, alcohol
    ]])
    quality = regressor.predict(sample)[0]
    display(f'Quality = {quality}')

style = {'description_width': 'initial'}

display(interactive(
    predict,
    wine_type=Dropdown(options=['White', 'Red'], style=style),
    fixed_acidity=FloatSlider(value=7, min=4.0, max=16.0, style=style),
    volatile_acidity=FloatSlider(value=0.35, min=0.0, max=1.6, style=style),
    citric_acid=FloatSlider(value=0.35, min=0.0, max=1.7, style=style),
    residual_sugar=IntSlider(value=5, min=1, max=66, style=style),
    chlorides=FloatSlider(value=0.05, min=0.0, max=0.65, style=style),
    free_sulfur_dioxide=IntSlider(value=30, min=1.0, max=290, style=style),
    other_sulfur_dioxide=IntSlider(value=86, min=5.0, max=260, style=style),
    density=FloatSlider(value=0.99, min=0.95, max=1.15, style=style),
    ph=FloatSlider(value=3, min=2.5, max=16, style=style),
    sulphates=FloatSlider(value=0.5, min=0.2, max=2, style=style),
    alcohol=FloatSlider(value=10.5, min=3.0, max=15.0, style=style)
))
```
